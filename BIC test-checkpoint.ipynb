```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import sklearn.datasets
from scipy.spatial import distance
import matplotlib.pyplot as plt
import pandas as pd

data =  np.loadtxt(fname = 'prep5.csv', delimiter = ',')

time = data[:, 0]   # execution time data
task = data[:, 1]   # number of task
```

```python
sse = []
kmean_value = []
list_k = list(range(3, 20))

for k in list_k:
    kmeans = KMeans(n_clusters=k, init="k-means++", random_state=0).fit(time.reshape(-1, 1))
    sse.append(kmeans.inertia_)
    kmean_value.append(kmeans)
    print("iteration: " + str(k))
```

```python
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of Squared Error(SSE)')
```

```python
BIC = []
for kmeani in kmean_value:
    centers = kmeani.cluster_centers_
    labels = kmeani.labels_ # 각 cluster의 data수
    
    #number of cluster
    m = kmeani.n_clusters
    #size of the cluster
    n = np.bincount(labels)
    #size of data set
    N = time.shape[0]

    
    #compute variance for all clusters beforehand
    cl_var = [(1.0 / (N - m)) * sum(abs(time[np.where(labels == i)] - centers[i])) for i in range(m)]
    const_term = 2 * m * np.log(N)

    #print(np.sum([n[i]* np.log(n[i]) - n[i]* np.log(N) - (n[i] / 2 ) * np.log(2*np.pi)  - (n[i] / 2) * np.log(cl_var[i]) - ((n[i] - m) / 2)for i in range(m)]))
    BIC.append( np.sum([n[i]* np.log(n[i]) - n[i]* np.log(N) - (n[i] / 2 ) * np.log(2*np.pi) - (n[i] / 2) * np.log(cl_var[i]) - ((n[i] - m) / 2)for i in range(m)]) - const_term)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(list_k, BIC, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('BIC')
```
